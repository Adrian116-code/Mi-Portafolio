```python
import pandas as pd
import numpy as np

data = {
    'Fecha': pd.date_range(start='2023-01-01', periods=10, freq='M'),
    'Producto': ['Laptop', 'Mouse', 'Monitor', 'Teclado', 'Laptop', 'Mouse', 'Monitor', 'Teclado', 'Laptop', 'Mouse'],
    'Ventas': [1200, 300, 800, 450, 1500, 350, 850, 500, 1300, 400],
    'Region': ['Norte', 'Sur', 'Norte', 'Este', 'Oeste', 'Norte', 'Sur', 'Este', 'Oeste', 'Norte']
}

df = pd.DataFrame(data)
df.to_csv('ventas_mini.csv', index=False)
print("Base de datos 'ventas_mini.csv' creada con éxito.")
```

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

df_dashboard = pd.read_csv('ventas_mini.csv')

fig_linea = px.line(df_dashboard, x='Fecha', y='Ventas', title='Evolución Mensual de Ventas',
              markers=True, template='plotly_dark')

fig_barras = px.bar(df_dashboard, x='Producto', y='Ventas', color='Region',
               title='Ventas por Producto y Región', template='plotly_dark')

fig_pastel = px.pie(df_dashboard, values='Ventas', names='Region',
               title='Distribución de Ventas por Región', hole=0.4, template='plotly_dark')

fig_linea.show()
fig_barras.show()
fig_pastel.show()
```

```python
fig = make_subplots(rows=1, cols=2,
                    specs=[[{"type": "bar"}, {"type": "pie"}]],
                    subplot_titles=("Ventas por Producto", "Cuota por Región"))

fig.add_trace(go.Bar(x=df_dashboard['Producto'], y=df_dashboard['Ventas'], name="Productos"), row=1, col=1)

fig.add_trace(go.Pie(labels=df_dashboard['Region'], values=df_dashboard['Ventas'], name="Regiones"), row=1, col=2)

fig.update_layout(height=500, title_text="Panel de Control de Ventas v1.0", showlegend=True, template='plotly_dark')
fig.show()
```

```python
fig.write_html("index.html")

from google.colab import files
files.download('index.html')
```
